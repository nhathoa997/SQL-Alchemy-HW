```python
def is_prime(n):
    """
    Given an integer n >= 1, this returns
    true if n is prime, and false if n is
    composite.
    """
    prime = True
    if abs(n) == 1: prime = False
    else:
        for i in range(2,abs(n)):
            if n%i == 0:
                prime = False
                return prime
            else:
                prime = True
    return prime
def prime_divisors(n):
    prime_divisors = []
    if abs(n) == 1: 
        prime_divisors.append(0)
        return prime_divisors
    elif is_prime(abs(n)) == True:
            prime_divisors.append(n)
            return prime_divisors
    else: 
        for i in range(2,abs(n)):
            if (n%i == 0 and is_prime(i) == True):
                prime_divisors.append(i)
        return prime_divisors
```

```python
def guess_key_length(C):
    #should return an integer which is the most likely keylength of the key. 
    #the for loop goes through each trigraph in the CT to see how many occurences that each trigraph has within the CT.
    #One looks for repeated fragments within the ciphertext and compiles a list of the distances that separate the repetitions. 
    #The key length is likely to divide many of these distances
    #trigraphs = []
    possible_key_lengths = []
    for i in range(len(C) - 3):
        distances = []
        trigraph = C[i:i+3]
        for j in range(len(C) - 3):
            test_trigraph = C[j:j+3]
            if trigraph == test_trigraph:
                distance = j - i
                if distance == 0:
                    continue
                else:
                    distances.append(prime_divisors(abs(j - i)))
        if len(distances) == 0:
            continue
        else:
            info = list(chain(*distances))
            possible_key_lengths.append(info)
    possible_key_lengths = list(chain(*possible_key_lengths))
    #Look for the most repeated number in the list of possible_key_lengths
    old_count = 0
    for k in range(len(possible_key_lengths)):
        if (possible_key_lengths.count(possible_key_lengths[k]) > old_count):
            old_count = possible_key_lengths.count(possible_key_lengths[k])
            key_length = possible_key_lengths[k]
    
    return key_length 
```

```python
def chunks(cipher, n):
    grouped_list = []
    for j in range(0,n):
        evenly_spaced_letters = []
        evenly_spaced_letters.append(cipher[j::n])
        grouped_list.append(evenly_spaced_letters)
    return grouped_list
```

```python
def most_common_letters(grouped_cipher):
    
    most_common_let = []
    for i in range(len(grouped_cipher)):
        count_per_let = []
        group_i = ''.join(grouped_cipher[i])
        for j in range(len(group_i)):
            letter = group_i[j]
            count = group_i.count(letter)
            info = (letter,count)
            count_per_let.append(info)
        count_per_let = list(OrderedDict.fromkeys(count_per_let))
        most_common_let.append(count_per_let)
    return most_common_let 
```

```python
def letter_frequency(cipher_count,grouped_cipher):
    letter_freq_per_group = []
    for i in range(len(grouped_cipher)):
        group_i = cipher_count[i]
        str_i = ''.join(grouped_cipher[i])
        letter_frequency = []
        for i in range(len(group_i)):
            letter = group_i[i][0]
            count = group_i[i][1]
            letter_freq = round((count/len(str_i)),3)
            info = (letter, letter_freq)
            letter_frequency.append(info)
        letter_freq_per_group.append(letter_frequency)
    return letter_freq_per_group
```

```python
def comparison(letter_frequency):
    most_likely_shift = []
    for x in range(len(letter_frequency)):
        probability = []
        freq_i = letter_frequency[x]
        for i in range(len(freq_i)):
            summation = 0
            for j in range(len(freq_i)):
                summation += freq_i[(j-i)%len(freq_i)][1]*english_let_freq(freq_i[j][0])
            shift = i
            info = (shift,summation)
            probability.append(info)
        for k in range(len(probability)):
            if probability[k][1] == max([x[1] for x in probability]):
                most_likely_shift.append(probability[k][0])
    return most_likely_shift
```
